## Geojson_hcmc

### Import libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
os.getcwd()
```

### Read data of district boundaries of Ho Chi Minh City

```python
DataFolder = 'data'
DataFile = 'district-boundary-hcm-city.geojson'
DataPath = os.path.join(DataFolder, DataFile)
DataPath
```

```python
# Read data file of district boundaries of Ho Chi Minh City (district-boundary-hcm-city.geojson)
district = gpd.read_file(DataPath)
district.head()
```

### Plot district boundaries of Ho Chi Minh City

```python
# Display the map of boundaries from district data, cmap = 'Set2'
plt.figure(figsize=(6,6))
district.plot(column = 'name', cmap = 'Set2', figsize=(16,12))
plt.show()
```

### Read data file of location of Ho Chi Minh City

```python
df_hcm = pd.read_excel('data/HCMC_location.xlsx', sheet_name='Location')
df_hcm.head()
```

#### Plot the map of Ho Chi Minh City with the location of districts

```python
plt.figure(figsize=(8,8))
plt.scatter(df_hcm.Longitude, df_hcm.Latitude)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title('HCMC districts Longitude vs Latitude')
plt.show()
```

#### Plot the map of HCMC with the location of each district

```python
plt.figure(figsize=(14,10))
district.plot(column = 'name', cmap = 'Set2', legend=True, figsize=(16,12))
plt.scatter(df_hcm.Longitude, df_hcm.Latitude)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title('Districts in HoChiMinh in City')
plt.grid()
plt.show()
```
